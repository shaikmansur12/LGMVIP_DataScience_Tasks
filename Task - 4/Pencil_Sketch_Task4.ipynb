# Task 4 :Image To Pencil Sketch  

### Importing cv2


```python
import cv2
```

### Reading The Image

```python
img=cv2.imread('C:/Users/mansurshaik/Downloads/animals.jpg')
cv2.imshow("Dogs",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Converting The Image Into GrayScale

```python
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("Dogs",new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Converting The GrayScaled Image To Inverted Image

```python
inver_img=255-new_img
cv2.imshow("Inverted Image",inver_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Making The Blur Image

```python
blur_img=cv2.GaussianBlur(inver_img,(21,21),0)
cv2.imshow("Blurred Image",blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Inverting The Blur Image And Convert Image To Pencil Sketch 

```python
inver_blur=255-blur_img
pencil_sketched=cv2.divide(new_img,inver_blur,scale=256.0)
cv2.imshow("Pencil_Sketched",pencil_sketched)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Comparing The original And Sketched Image

```python
cv2.imshow("Original Pic",img)
cv2.imshow("Pencil Sketched Pic",pencil_sketched)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
